```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
tips = sns.load_dataset("tips")
```

```python
# Grouping the data
daily_tips = tips.groupby("day")["tip"].sum()

# Plotting
plt.figure(figsize=(8, 5))
plt.bar(daily_tips.index, daily_tips.values)

# Adding labels and title
plt.xlabel("Day of the Week")
plt.ylabel("Total Tip Amount")
plt.title("Total Tips Collected per Day")

plt.show()
```

```python
# Creating the histogram
sns.histplot(tips['tip'], bins=20, kde=True)

# Adding title
plt.title("Distribution of Tip Amounts")

plt.show()


#"The histogram shows that most tips fall between $2 and $4, making the distribution slightly right-skewed. 
#The KDE line helps visualize the smooth probability density, highlighting that larger tips (over $6) are much less frequent."
```

```python
# Creating the scatter plot
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")

# Adding title
plt.title("Total Bill vs Tip Amount")

plt.show()

#"There is a clear positive correlation between the total bill and the tip amount; as the bill increases, the tip generally increases as well."
```
